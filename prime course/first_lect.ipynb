```python
import json
```

```python
def load_data(filename):
    with open(filename,'r') as f:
        data =json.load(f)
    return data
```

```python
def load_lines(filename):
    with open(filename,'r') as f:
        data =json.load(f)
    for line in data:
        print(line)
```
