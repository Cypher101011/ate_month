```python
import pandas as pd

df = pd.read_csv("raw_data.csv")

df
```

```python
df.groupby("country")["income"].mean()
```

```python
df.groupby("country")["income"].max()
```

```python
df.groupby("country")["income"].min()
```

```python
df.groupby("country")["income"].agg(["mean","min","max"])
```

```python
df.groupby("country")["income"].aggregate(avg_salary="mean",min_salary="min",max_salary="max")
```

```python
df.groupby("country")["income"].agg(avg_salary="mean",min_salary="min",max_salary="max")
```

```python
df.groupby("country").agg({
    "income":"mean",
    "age":"mean"})
```

```python
df.groupby("country").agg(
    
    max_salary=("income","max"),
   age=("age","mean"))
```

```python
df =pd.DataFrame({
    "country":['USA','USA','India','India'],
    'year':[2020,2021,2020,2021],
    'sales':[100,120,90,110],
    'profit':[20,25,18,22],
    'tax':[29,22,11,13]
})
melted_df=df.melt(
    id_vars=['country','year'],
    value_vars=['sales','profit','tax'],
    var_name="metrix",
    value_name="value"
)
```

```python
melted_df
```

```python
orignal=melted_df.pivot(
    index=["country",'year'],
    columns="metrix",
    
    values="value"
)
```

```python
orignal
```

```python
df_customers=pd.DataFrame({
    "customer_id":[1,2,3,4],
    'name':['adam','bob','charlie','dave']
})
df_orders=pd.DataFrame({
    'order_id':[101,102,103,104],
    'customer_id':[2,1,3,5],
    'amount':[240,120,300,180],
})
```

```python
pd.merge(df_customers,df_orders,on='customer_id')
pd.merge(df_customers,df_orders,on='customer_id',how='left')
```

```python
pd.merge(df_customers,df_orders,on='customer_id',how='right')
```

```python
pd.merge(df_customers,df_orders,on='customer_id',how='outer')
```
