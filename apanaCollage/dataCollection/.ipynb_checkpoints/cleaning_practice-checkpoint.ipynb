```python
df["Quantity_num"]=pd.to_numeric(df["Quantity"],errors='coerce')
```

```python
df["Price_num"]=pd.to_numeric(df["Price Per Unit"],errors='coerce')
```

```python
df["Total_num"]=pd.to_numeric(df["Total Spent"],errors="coerce")
```

```python
df[["Quantity", "Quantity_num", "Price Per Unit", "Price_num", "Total Spent", "Total_num"]].head(100)
```

```python
df.head()
```

```python
for col in ['Item','Payment Method','Location']:
    df[col]=(df[col]
        .str.strip()
        .str.lower()
        .replace({"error":None,"unknown":None,"":None})
                )
```
